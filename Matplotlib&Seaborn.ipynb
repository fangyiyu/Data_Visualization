```python
# Add annotation
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure(figsize=(15,10))
ax=fig.add_subplot(1,1,1)
data=pd.read_csv('Main form historical data from 1981.csv',index_col=0, parse_dates=True)
gold_price=data['GOLD']
gold_price.plot(ax=ax, style='k-')

crisis_data=[(datetime(2001,9,11),'911'),(datetime(2008,3,12),'Financial Crisis')]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, gold_price.asof(date)+50),
                xytext=(date, gold_price.asof(date)+300),
                arrowprops=dict(facecolor='black', headwidth=4, width=2, headlength=8),
                horizontalalignment='left', verticalalignment='top'
               )
ax.set_xlim(['3/1/2001','10/24/2019'])
ax.set_ylim([100,2000])
ax.set_title('Important events in Gold Price')
plt.savefig('Events.png',dpi=400,bbox_inches='tight')
plt.show()  
```

```python
tips=pd.read_csv('tips.csv')
tips.head()
```

```python
party_counts=pd.crosstab(tips['day'],tips['size'])
tips['tip_pct']=tips['tip']/(tips['total_bill']-tips['tip'])
```

```python
party_counts
```

```python
party_pcts=party_counts.div(party_counts.sum(1),axis=0)
party_pcts
```

```python
party_pcts.plot.bar()
```

```python
tips['tip_pct'].plot.hist(bins=50)
```

```python
tips['tip_pct'].plot.density()
```

```python
macro=pd.read_csv('macrodata.csv')
data=macro[['cpi','m1','tbilrate','unemp']]
tran_data=np.log(data).diff().dropna()
```

```python
tran_data.head(5)
```

```python
fig=plt.figure()
plt.scatter(tran_data['m1'],tran_data['unemp'])
fig.show()
```

```python
import seaborn as sb
fig=plt.figure()
sb.regplot('m1','unemp',data=tran_data,marker='+',ci=68)
plt.show()
```
